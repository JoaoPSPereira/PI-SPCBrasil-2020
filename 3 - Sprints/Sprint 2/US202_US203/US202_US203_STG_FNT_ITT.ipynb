```python
import pandas as pd
low_memory=False
%matplotlib inline
import warnings; warnings.simplefilter("ignore")
```

```python
filepath = './base'
filename = '/STG_FNT_ITT.xlsx'
df = pd.read_excel(filepath + filename, error_bad_lines=False, encoding='latin-1')
```

```python
##Exibindo o dataframe da tabela
df
```

```python
##Exibindo lista de colunas
df.columns
```

```python
##renomeando as colunas
df = df.rename(columns={'ID_STG_FNT_ITT': 'Id', 'NUM_CNPJ': 'cnpj', 'NUM_CMP_CNPJ': 'cmp_cnpj', 'NOM_COM': 'nome_comercial',
                        'NOM_RAZ_SCL': 'nome_razao_social', 'DAT_INC_DBO': 'data_de_inclusao'})
##Ordenando as linhas pelo Id em ordem crescente
df.sort_values('Id')
```

```python
##VerificaÃ§Ã£o da quantidade de caracteres

##transformar as colunas em str
df['cnpj'] = df['cnpj'].astype('str')
df['cmp_cnpj'] = df['cmp_cnpj'].astype('str')

##Deixar no dataframe somente as linhas que possuem cpf com numero de digitos diferentes de 14
df_correct = df[df['cnpj'].str.len() + df['cmp_cnpj'].str.len() == 14]
df_error = df[df['cnpj'].str.len() + df['cmp_cnpj'].str.len() != 14]

df1 = df[df['cnpj'].str.len() + 1 + df['cmp_cnpj'].str.len() == 14]
df2 = df[df['cnpj'].str.len() + 2 + df['cmp_cnpj'].str.len() == 14]
```

```python
df1['cnpj'] = '0' + df1['cnpj']
df2['cnpj'] = '00' + df2['cnpj']

df_correct = pd.concat([df_correct,df1])
df_correct = pd.concat([df_correct,df2])

df_correct
```

```python
df_error = df_error[~df_error.Id.isin(df_correct.Id)]
df_error
df_correct
```

```python
## Transformar as colunas cnpj e cmp_cnpj em int
df_correct['cnpj'] = df['cnpj'].astype('int64')
df_correct['cmp_cnpj'] = df['cmp_cnpj'].astype(int)
df_error['cnpj'] = df['cnpj'].astype('int64')
df_error['cmp_cnpj'] = df['cmp_cnpj'].astype(int)

## Renomear colunas novamente para a exportaÃ§Ã£o
df_correct = df_correct.rename(columns={'Id': 'ID_STG_FNT_ITT', 'cnpj': 'NUM_CNPJ', 'cmp_cnpj': 'NUM_CMP_CNPJ', 'nome_comercial' : 'NOM_COM',
                        'nome_razao_social': 'NOM_RAZ_SCL', 'data_de_inclusao':  'DAT_INC_DBO'})
df_error = df_error.rename(columns={'Id': 'ID_STG_FNT_ITT', 'cnpj': 'NUM_CNPJ', 'cmp_cnpj': 'NUM_CMP_CNPJ', 'nome_comercial' : 'NOM_COM',
                        'nome_razao_social': 'NOM_RAZ_SCL', 'data_de_inclusao':  'DAT_INC_DBO'})

## Exportar df para xlsx ordenado pelo ID_STG_FNT_ITT em ordem crescente e sem index
df_correct.sort_values('ID_STG_FNT_ITT').to_excel("Linhas_Corretas_STG_FNT_ITT.xlsx", index=False)
df_error.sort_values('ID_STG_FNT_ITT').to_excel("Linhas_Incorretas_STG_FNT_ITT.xlsx", index=False)
```
